```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load your dataset
df = pd.read_csv("weatherHistory.csv")  # Update path if needed
def clean_data(dataset):

  dataset.dropna(inplace=True)

  return dataset

df= clean_data(df)
df.rename(columns={"Formatted Date": "DateTime"}, inplace=True)
df["DateTime"] = pd.to_datetime(df["DateTime"], utc=True, errors='coerce')

# Convert to datetime and extract parts
df['DateTime'] = pd.to_datetime(df['DateTime'], utc=True)
df['hour'] = df['DateTime'].dt.hour
df['day'] = df['DateTime'].dt.day
df['month'] = df['DateTime'].dt.month

# Drop rows with nulls
df.dropna(inplace=True)

# Select input features (datetime-based) and output features (weather values)
X = df[['hour', 'day', 'month']].values
y = df[['Temperature (C)','Humidity', 'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)', 'Loud Cover', 'Pressure (millibars)']].values
```

```python
df
```

```python
# Scale inputs and outputs
x_scaler = StandardScaler()
X_scaled = x_scaler.fit_transform(X)

y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


model = Sequential([
    Dense(64, activation='relu', input_shape=(X_scaled.shape[1],)),
    Dense(64, activation='relu'),
    Dense(7)  # 7 outputs
])

model.compile(optimizer='adam', loss='mse')

model.fit(X_scaled, y_scaled, epochs=50, batch_size=32, validation_split=0.2)
```

```python
def predict_weather_from_datetime(input_datetime_str):
    import datetime
    dt = pd.to_datetime(input_datetime_str, utc=True)

    # Create input features: hour, day, month
    input_features = np.array([[dt.hour, dt.day, dt.month]])
    input_scaled = x_scaler.transform(input_features)

    # Predict using the trained model
    pred = model.predict(input_scaled)[0]

    # Inverse transform the scaled prediction
    numeric_pred = y_scaler.inverse_transform(pred.reshape(1, -1))[0]

    # Map predicted values to labels
    print(f"\n📅 Predicted Weather on {input_datetime_str}:")
    print(f"🌡️  Temperature: {numeric_pred[0]:.2f} °C")
    print(f"💧 Humidity: {numeric_pred[1]*100:.2f}%")
    print(f"💨 Wind Speed: {numeric_pred[2]:.2f} km/h")
    print(f"🧭 Wind Direction: {numeric_pred[3]:.2f}°")
    print(f"👁️ Visibility: {numeric_pred[4]:.2f} km")
    print(f"☁️ Cloud (Loud) Cover: {numeric_pred[5]*100:.2f}%")
    print(f"📉 Pressure: {numeric_pred[6]:.2f} millibars")
```

```python
import requests
import datetime

def get_real_time_weather(city_name, api_key):
    base_url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        'q': city_name,
        'appid': api_key,
        'units': 'metric' # Gets temp in Celsius, wind speed in m/s
    }

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status() # Raise an exception for bad status codes (4xx or 5xx)
        data = response.json()

        # Extract relevant fields, handling potential missing keys gracefully
        weather_data = {
            'Temperature': data['main'].get('temp'),
            'Apparent Temperature': data['main'].get('feels_like'), # Get 'feels_like' too
            'Humidity': data['main'].get('humidity'),
            'Wind Speed': data['wind'].get('speed'),
            'Wind Direction': data['wind'].get('deg'),
            'Pressure': data['main'].get('pressure'),
            'Visibility': data.get('visibility', 10000) / 1000,
            'Cloud Cover': data['clouds'].get('all'),
            'Summary': data['weather'][0].get('description') if data.get('weather') else None,
            'Precip Type': None,
            'Loud Cover': 0.0,
            'Daily Summary': None
        }
        # Basic precipitation type mapping (optional improvement)
        if weather_data['Summary']:
            summary_lower = weather_data['Summary'].lower()
            if 'rain' in summary_lower or 'drizzle' in summary_lower:
                 weather_data['Precip Type'] = 'rain'
            elif 'snow' in summary_lower or 'sleet' in summary_lower:
                 weather_data['Precip Type'] = 'snow'
            # Add other mappings if needed (hail, etc.)

        return weather_data

    except requests.exceptions.RequestException as e:
        print(f"❌ Error fetching real-time data for {city_name}: {e}")
        if response is not None:
             print(f"❌ API Response Status: {response.status_code}")
             print(f"❌ API Response Body: {response.text}")
        return None
    except KeyError as e:
        print(f"❌ Error parsing real-time data for {city_name}: Missing key {e}")
        print(f"❌ Received data structure: {data}")
        return None
    except Exception as e: # Catch other potential errors during parsing
        print(f"❌ An unexpected error occurred parsing data for {city_name}: {e}")
        return None
```

```python
def append_real_weather_to_csv(city_name, api_key, csv_filepath="weatherHistory.csv"):
    """
    Fetches real-time weather for a city and appends it to the specified CSV file,
    using ONLY the columns specified by the user image.

    Args:
        city_name (str): The name of the city to get weather for (NOT added to CSV).
        api_key (str): Your OpenWeatherMap API key.
        csv_filepath (str): The path to the weather history CSV file.
    """
    print(f"Fetching real-time weather for {city_name}...")
    real_time_data = get_real_time_weather(city_name, api_key)

    if real_time_data is None:
        print(f"Could not fetch or parse data. CSV not updated.")
        return

    # Get the current time in UTC
    current_utc_time = pd.Timestamp.now(tz='UTC')
    # Format to match the likely original CSV format string (adjust if needed)
    formatted_date_string = current_utc_time.strftime('%Y-%m-%d %H:%M:%S.%f %Z')

    # Define the exact order and names from your image
    column_order = [
        'Formatted Date', 'Summary', 'Precip Type', 'Temperature (C)',
        'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)',
        'Wind Bearing (degrees)', 'Visibility (km)', 'Loud Cover',
        'Pressure (millibars)', 'Daily Summary'
    ]

    # Prepare the data row dictionary using the EXACT column names from the image
    new_row_data = {
        'Formatted Date': formatted_date_string,
        'Summary': real_time_data.get('Summary', np.nan),
        'Precip Type': real_time_data.get('Precip Type', np.nan), # Will be 'rain'/'snow'/nan based on logic above
        'Temperature (C)': real_time_data.get('Temperature'),
        'Apparent Temperature (C)': real_time_data.get('Apparent Temperature'),
        'Humidity': real_time_data.get('Humidity') / 100.0 if real_time_data.get('Humidity') is not None else np.nan, # Convert % to 0-1 scale
        'Wind Speed (km/h)': real_time_data.get('Wind Speed') * 3.6 if real_time_data.get('Wind Speed') is not None else np.nan, # Convert m/s to km/h
        'Wind Bearing (degrees)': real_time_data.get('Wind Direction'),
        'Visibility (km)': real_time_data.get('Visibility'),
        'Loud Cover': real_time_data.get('Loud Cover', 0.0), # Use default 0.0
        'Pressure (millibars)': real_time_data.get('Pressure'), # hPa is equivalent to millibar
        'Daily Summary': real_time_data.get('Daily Summary', np.nan) # Not available from current API
    }

    # Create a DataFrame for the new row using the specific column order
    # This ensures only these columns are included, in this exact order.
    new_row_df = pd.DataFrame([new_row_data], columns=column_order)

    # Check for potential missing values (optional, for information)
    if new_row_df.isnull().any().any():
        print("Warning: Some weather data fields might be missing from the API and are stored as NaN.")
        # print(new_row_df.isnull().sum()) # Uncomment to see specific missing counts


    # Append the new row to the CSV file without writing the header again
    try:
        # Check if file exists to determine if header needs writing
        file_exists = False
        try:
            # Try reading just the header; will fail if file doesn't exist
            pd.read_csv(csv_filepath, nrows=0)
            file_exists = True
        except FileNotFoundError:
            file_exists = False
            print(f"CSV file '{csv_filepath}' not found. Creating a new file.")
        except Exception as e:
            print(f"Warning: Could not properly check if CSV '{csv_filepath}' exists. Appending without header. Error: {e}")
            # Assume it exists to avoid duplicate headers if unsure

        new_row_df.to_csv(csv_filepath, mode='a', header=(not file_exists), index=False)
        if file_exists:
             print(f"Successfully appended real-time weather for {city_name} to {csv_filepath}")
        else:
             print(f"Successfully created {csv_filepath} with real-time weather for {city_name}")

    except Exception as e:
        print(f"❌ Error writing data to CSV '{csv_filepath}': {e}")

# --- Example Usage (Place this at the end of your script) ---
# Make sure API_KEY is defined earlier in your script
# API_KEY = "YOUR_API_KEY" # Replace with your actual key

TARGET_CITY = "Pune" # The city you want the current weather for
CSV_FILE = "weatherHistory.csv" # Path to your CSV

append_real_weather_to_csv(TARGET_CITY, API_KEY, CSV_FILE)

print("\nScript finished.")
```

```python
def compare_prediction_with_real_time(input_datetime_str, city_name, api_key):
    # Prediction
    dt = pd.to_datetime(input_datetime_str, utc=True)
    input_features = np.array([[dt.hour, dt.day, dt.month]])
    input_scaled = x_scaler.transform(input_features)
    pred = model.predict(input_scaled)[0]
    numeric_pred = y_scaler.inverse_transform(pred.reshape(1, -1))[0]

    print(f"\n🔮 Predicted Weather on {input_datetime_str} for {city_name}:")
    print(f"🌡️ Temperature: {numeric_pred[0]:.2f} °C")
    print(f"💧 Humidity: {numeric_pred[1]*100:.2f}%")
    print(f"💨 Wind Speed: {numeric_pred[2]:.2f} km/h")
    print(f"🧭 Wind Direction: {numeric_pred[3]:.2f}°")
    print(f"👁️ Visibility: {numeric_pred[4]:.2f} km")
    print(f"☁️ Cloud (Loud) Cover: {numeric_pred[5]*100:.2f}%")
    print(f"📉 Pressure: {numeric_pred[6]:.2f} millibars")

    # Real-time
    real = get_real_time_weather(city_name, api_key)
    if real:
        print(f"\n🌍 Real-Time Weather Now in {city_name}:")
        print(f"🌡️ Temperature: {real['Temperature']} °C")
        print(f"💧 Humidity: {real['Humidity']}%")
        print(f"💨 Wind Speed: {real['Wind Speed']} m/s")
        print(f"🧭 Wind Direction: {real['Wind Direction']}°")
        print(f"👁️ Visibility: {real['Visibility']} km")
        print(f"☁️ Cloud Cover: {real['Cloud Cover']}%")
        print(f"📉 Pressure: {real['Pressure']} hPa")
```
